# A Map of Apartments, Pet related services and goods, Parks & Trails, and Grocery stores

```python
venues_map = folium.Map(width=900, height=800,location=[latitude, longitude], zoom_start=15) # generate map centred around central downtown with 2 mile / 3218.69 meter radius of Downtown

# add a red circle marker to represent the center point
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Central downtown',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

# add the Apartments as circle markers
for lat, lng, label in zip(apartments_dataframe_filtered.lat, apartments_dataframe_filtered.lng, apartments_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='magenta',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

# add the Pet related services and stores as blue circle markers
for lat, lng, label in zip(pet_dataframe_filtered.lat, pet_dataframe_filtered.lng, pet_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=6,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)
    
# add the groceries as circle markers green and lime
for lat, lng, label in zip(grocery_dataframe_filtered.lat, grocery_dataframe_filtered.lng, grocery_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=4,
        color='green',
        popup=label,
        fill = True,
        fill_color='lime',
        fill_opacity=0.6
    ).add_to(venues_map)

# add the groceries as circle markers green and lime
for lat, lng, label in zip(trail_dataframe_filtered.lat, trail_dataframe_filtered.lng, trail_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=8,
        color='yellow',
        popup=label,
        fill = True,
        fill_color='brown',
        fill_opacity=0.6
    ).add_to(venues_map)

# display map
venues_map
```

# Analyzing Apartments, the priority criteria for *Paw & Claw*

```python
# one hot encoding
apartments_onehot = pd.get_dummies(apartments_dataframe_filtered[['categories']], prefix="", prefix_sep="")

# add distance column back to dataframe
apartments_onehot['name'] = apartments_dataframe_filtered['name'] 

# move neighborhood column to the first column
fixed_columns = [apartments_onehot.columns[-1]] + list(apartments_onehot.columns[:-1])
apartments_onehot = apartments_onehot[fixed_columns]

apartments_onehot.head()
```

```python
apartments_onehot.shape
```

```python
apartments_grouped = apartments_onehot.groupby('name').mean().reset_index()
apartments_grouped.head()
```

```python
apartments_grouped.shape
```

```python
num_top_venues = 10

for hood in apartments_grouped['name']:
    print("----"+hood+"----")
    temp = apartments_grouped[apartments_grouped['name'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```
